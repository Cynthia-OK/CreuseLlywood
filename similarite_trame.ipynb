```python
import pandas as pd
films = pd.read_csv('./donnees/export_films.csv', sep='\t', low_memory=False)
films_personnes = pd.read_csv('./donnees/export_films_personnes.csv',sep='\t', low_memory=False)
```

on explode les colonnes en utilisant la fonction eval

```python
import ast

def transfo_liste(x):
  if isinstance(x, str):
    return ast.literal_eval(x)
  else:
    return x
```

```python
films['genre_liste'] = films['genres_x'].apply(transfo_liste)
films.head()
```

```python
tous_les_genres = set()
for genres in films['genre_liste']:
  tous_les_genres.update(genres)

tous_les_genres
```

```python
for genre in tous_les_genres:
  films[f'genre_{genre}'] = films['genre_liste'].apply(lambda x: genre in x)

films.head()
```

```python
films = films[(films['genres_x'].str.contains('Movie')==False)]
films
```
